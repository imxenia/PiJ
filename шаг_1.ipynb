```
import re
```

```
import numpy as np
```

```
def clear_runtime(runtime):
    return re.sub(r'\D', '', runtime)
```

```
dataset['runtime_clear'] = dataset['runtime'].apply(clear_runtime)
```

```
dataset['runtime_clear']
```

```
dataset['runtime_clear'].isna().sum()
```

```
dataset = dataset[dataset['runtime_clear'] != '']
```

```
dataset.shape
```

```
dataset.info()
```

```
dataset['runtime_clear'] = dataset['runtime_clear'].astype(int)
```

```
dataset.info()
```

```
dataset['gross_earn']
```

```
def clear_gross_earn(gross_earn):
    if pd.isna(gross_earn):
        return np.nan
    else:
        return re.sub('[^\d.]', '', gross_earn)
```

```
dataset['gross_earn_clear'] = dataset['gross_earn'].apply(clear_gross_earn)
```

```
dataset['gross_earn_clear']
```

```
dataset['gross_earn_clear'] = dataset['gross_earn_clear'].astype(float)
```

```
dataset.info()
```

```
dataset['gross_earn_clear']
```

```
sorted_dataset = dataset.sort_values(by='rating', ascending=False)
```

С помощью dataset.shape можно понять, сколько строк и столбцов в датасете. В данном случае количество  наблюдений - 5000, количество признаков - 11.

В датасете присутствуют следующие типы данных:
object (строки) и float64.

Предобработка типов данных:
Признаки runtime, votes, и gross_earn представлены как строки (object), хотя это числовые значения. Эти столбцы нужно привести к числовому типу, чтобы с ними можно было проводить арифметические операции.
Также нужно убедиться, что нет лишних символов в этих столбцах, таких как "часов" или "минут" в runtime, которые нужно будет удалить для корректной работы с данными.

Дубликаты:
В датасете нет дубликатов, что подтверждается результатом dataset.duplicated().sum(), равным 0.

Предобработка:
Так как дубликатов нет, дополнительная обработка на этом этапе не требуется.

Пропущенные значения:
В столбце genre отсутствуют данные для 30 фильмов (5000 - 4970 = 30).
В столбце about отсутствуют данные для 14 фильмов (5000 - 4986 = 14).
В столбце gross_earn отсутствуют данные для 415 фильмов (5000 - 4585 = 415).

Предобработка пропусков:
Пропуски в столбце gross_earn составляют значительный процент (415 из 5000), что может существенно повлиять на анализ. Для их обработки можно использовать медиану или среднее значение, либо оставить пропуски, если они не критичны для конкретного анализа.
Пропуски в столбцах genre и about можно заполнить наиболее часто встречающимся значением или оставить, если это не оказывает значимого влияния на результаты.

О возможных проблемах, которые могут возникнуть при анализе:
Преобразование столбцов runtime, votes и gross_earn в числовые значения потребует предварительной очистки данных. Например, в столбце gross_earn могут быть строковые символы в значениях, которые нужно удалить.
Пропуски в столбце gross_earn могут быть связаны с отсутствием информации о доходах некоторых фильмов, что может повлиять на анализ по доходности.
Возможно, в столбцах runtime и votes будут присутствовать значения, которые не соответствуют ожидаемым формам. Эти проблемы могут потребовать дополнительной очистки.
В столбце genre могут быть указаны несколько жанров для одного фильма, что потребует дополнительной обработки для правильной работы с этим признаком, например, разделение жанров по запятой.
